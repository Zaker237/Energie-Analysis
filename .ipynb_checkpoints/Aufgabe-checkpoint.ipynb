```python
#hier wir importieren die Tools, die wir brauchen
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#hier werden die beiden Datei als Dataframe (df2016 und df2017) gelesen
df2016 = pd.read_csv("Actual Generation per Production Type_201601010000-201701010000.csv")
df2016['year'] = 2016
df2017 = pd.read_csv("Actual Generation per Production Type_201701010000-201801010000.csv")
df2017['year'] = 2017
```

```python
def extra_year(df):
    for index, data in df.iterrows():
        year = data.MTU[6:10]
        df['test'] = year
```

```python
#die beiden DataFrame werden hier untereinanders geschrieben und geschpeichen als ein Dataframe (df)
df = pd.concat([df2016, df2017])
print(len(df))
df[:]
```

```python
all_keys = ["""Biomass  - Actual Aggregated [MW]""", """Fossil Brown coal/Lignite  - Actual Aggregated [MW]""", 
            """Fossil Coal-derived gas  - Actual Aggregated [MW]""", """Fossil Gas  - Actual Aggregated [MW]""", """Fossil Hard coal  - Actual Aggregated [MW]""", 
            """Fossil Oil  - Actual Aggregated [MW]""", """Fossil Oil shale  - Actual Aggregated [MW]""", """Fossil Peat  - Actual Aggregated [MW]""", 
            """Geothermal  - Actual Aggregated [MW]""", """Hydro Pumped Storage  - Actual Aggregated [MW]""", """Hydro Pumped Storage  - Actual Consumption [MW]""", 
            """Hydro Run-of-river and poundage  - Actual Aggregated [MW]""", """Hydro Water Reservoir  - Actual Aggregated [MW]""", """Marine  - Actual Aggregated [MW]""", 
            """Nuclear  - Actual Aggregated [MW]""", """Other  - Actual Aggregated [MW]""", """Other renewable  - Actual Aggregated [MW]""", """Solar  - Actual Aggregated [MW]""", 
            """Waste  - Actual Aggregated [MW]""", """Wind Offshore  - Actual Aggregated [MW]""", """Wind Onshore  - Actual Aggregated [MW]"""]

wind_keys = ["""Solar  - Actual Aggregated [MW]""", """Wind Offshore  - Actual Aggregated [MW]""", """Wind Onshore  - Actual Aggregated [MW]"""]

```

```python
#df['Share Wind+PV'] = df.apply(lamda row: np.sum(row[wind_key].to_numpy()) / np.sum(row[all_keys].to_numpy()))
#df2016['Share Wind+PV'] = df2016[wind_keys].sum(axis=0, skipna=True) / df2016[all_keys].sum(axis=0, skipna=True)
#df2017['Share Wind+PV'] = df2017[wind_keys].sum(axis=0, skipna=True) / df2017[all_keys].sum(axis=0, skipna=True)
#df['Share Wind+PV'] = df[wind_keys].sum(axis=0, skipna=True) / df[all_keys].sum(axis=0, skipna=True)

#df['Share Wind+PV'] = df.apply(lamda row: row[wind_keys].sum(axis=1, skipna=True)/row[all_keys].sum(axis=1, skipna=True))
df2016['Share Wind+PV'] = df2016[wind_keys].sum(axis=1, skipna=True)*100 / df2016[all_keys].sum(axis=1, skipna=True)
df2017['Share Wind+PV'] = df2017[wind_keys].sum(axis=1, skipna=True)*100 / df2017[all_keys].sum(axis=1, skipna=True)
df['Share Wind+PV'] = df[wind_keys].sum(axis=1, skipna=True)*100 / df[all_keys].sum(axis=1, skipna=True)

#export die DataFrame als XLXS
#df.to_excel('Dataframe.xlsx', index=None, header=True)
writer = pd.ExcelWriter('Dataframe.xlsx', engine='xlsxwriter')
df.to_excel(writer)
writer.save()
```
